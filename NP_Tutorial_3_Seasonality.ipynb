```python
###Uma Analise de como as Sazonalidades impactam no modelo e como saber quando usalas
```

```python
import pandas as pd
from neuralprophet import NeuralProphet, set_log_level
```

```python
#carregando df
df = pd.read_csv("https://github.com/ourownstory/neuralprophet-data/raw/main/kaggle-energy/datasets/tutorial01.csv")
```

```python
df
```

```python
# apenas habilitando a sazonalidade anual
m = NeuralProphet(
    n_changepoints=0,
    yearly_seasonality=True,
    weekly_seasonality=False,
    daily_seasonality=False)
m.set_plotting_backend("plotly")
metrics = m.fit(df)
forecast = m.predict(df)
m.plot(forecast)
```

```python
m.plot_components(forecast)
```

```python
m.plot_parameters(components=["seasonality"])
```

```python
#habilitando todas as sazonalidades
m = NeuralProphet(
    n_changepoints=0,
    yearly_seasonality=True,
    weekly_seasonality=True,
    daily_seasonality=True)
m.set_plotting_backend("plotly")
metrics = m.fit(df)
forecast = m.predict(df)
m.plot(forecast)
```

```python
m.plot_components(forecast, components=["seasonality"])
```

```python
m.plot_parameters(components=["seasonality"])
```
